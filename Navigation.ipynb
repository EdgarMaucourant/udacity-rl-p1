### 2. Examine the State and Action Spaces

The simulation contains a single agent that navigates a large environment.  At each time step, it has four actions at its disposal:
- `0` - walk forward 
- `1` - walk backward
- `2` - turn left
- `3` - turn right

The state space has `37` dimensions and contains the agent's velocity, along with ray-based perception of objects around agent's forward direction.  A reward of `+1` is provided for collecting a yellow banana, and a reward of `-1` is provided for collecting a blue banana. 

Run the code cell below to print some information about the environment.

````python
# reset the environment
env_info = env.reset(train_mode=True)[brain_name]

# number of agents in the environment
print('Number of agents:', len(env_info.agents))

# number of actions
action_size = brain.vector_action_space_size
print('Number of actions:', action_size)

# examine the state space 
state = env_info.vector_observations[0]
print('States look like:', state)
state_size = len(state)
print('States have length:', state_size)
````

### 3. Take Random Actions in the Environment

In the next code cell, you will learn how to use the Python API to control the agent and receive feedback from the environment.

Once this cell is executed, you will watch the agent's performance, if it selects an action (uniformly) at random with each time step.  A window should pop up that allows you to observe the agent, as it moves through the environment.  

Of course, as part of the project, you'll have to change the code so that the agent is able to use its experience to gradually choose better actions when interacting with the environment!

````python
env_info = env.reset(train_mode=False)[brain_name] # reset the environment
state = env_info.vector_observations[0]            # get the current state
score = 0                                          # initialize the score
while True:
    action = np.random.randint(action_size)        # select an action
    env_info = env.step(action)[brain_name]        # send the action to the environment
    next_state = env_info.vector_observations[0]   # get the next state
    reward = env_info.rewards[0]                   # get the reward
    done = env_info.local_done[0]                  # see if episode has finished
    score += reward                                # update the score
    state = next_state                             # roll over the state to next time step
    if done:                                       # exit loop if episode finished
        break
    
print("Score: {}".format(score))
````

### 4. It's Your Turn!

Now it's your turn to train your own agent to solve the environment!  When training the environment, set `train_mode=True`, so that the line for resetting the environment looks like the following:

First we define a method to train the Agent. This method is based on the Deep Q-Learning algorithm from the course. 
This method takes as input:
- The environement (based on ML Agents), 
- The Brain name (defined by the environment) 
- The agent to be trained
- The target average rewards we aim for to consider the agent successul. 
- Others parameters includes the number of episodes maximum that we train on and the number of steps maximum that we train the agent on, plus the epsilon values for the epsilon-greedy algorithm.

````python
def dqn(env, brain_name, agent, max_avg_reward, n_episodes=2000, max_t=1000, eps_start=1.0, eps_end=0.01, eps_decay=0.995):
    """Deep Q-Learning.
    
    Params
    ======
        n_episodes (int): maximum number of training episodes
        max_t (int): maximum number of timesteps per episode
        eps_start (float): starting value of epsilon, for epsilon-greedy action selection
        eps_end (float): minimum value of epsilon
        eps_decay (float): multiplicative factor (per episode) for decreasing epsilon
    """
    scores = []                        # list containing scores from each episode
    avg_scores = []
    scores_window = deque(maxlen=100)  # last 100 scores
    eps = eps_start                    # initialize epsilon
    for i_episode in range(1, n_episodes+1):
        env_info = env.reset(train_mode=True)[brain_name]
        state = env_info.vector_observations[0]
        score = 0
        last_avg_scores = 0
        for t in range(max_t):
            action = agent.act(state, eps)
            env_info = env.step(action)[brain_name]
            next_state = env_info.vector_observations[0]   # get the next state
            reward = env_info.rewards[0]                   # get the reward
            done = env_info.local_done[0]
            agent.step(state, action, reward, next_state, done)
            state = next_state
            score += reward
            if done:
                break 
        scores_window.append(score)       # save most recent score
        scores.append(score)              # save most recent score
        eps = max(eps_end, eps_decay*eps) # decrease epsilon
        print('\rEpisode {}\tAverage Score: {:.2f}'.format(i_episode, np.mean(scores_window)), end="")
        if i_episode % 100 == 0:
            last_avg_scores = np.mean(scores_window)
            avg_scores.append(last_avg_scores) 
            print('\rEpisode {}\tAverage Score: {:.2f}'.format(i_episode, last_avg_scores))
        if last_avg_scores>=max_avg_reward:
            print('\nEnvironment solved in {:d} episodes!\tAverage Score: {:.2f}'.format(i_episode-100, last_avg_scores))
            torch.save(agent.qnetwork_local.state_dict(), 'checkpoint.pth')
            break
    
    return scores, avg_scores
````

The first step is to create an agent to be trained. This agent is based on a model defined in the dqn_agent.py and dqn_model.py files. It is based on a simple Multi Layer Perceptron (MLP) with one hidden layer.

The agent takes has input the number of states in the game (37 as given by the ML-Agents), the number of actions (4) and a seed to randomize the initialization

Note that this model was trained locally on a Windows machine, I also used the Udacity workspace at first but the training was 10x slower so I ended up using my local machine instead. It has been trained on CPU only.

````python
from unityagents import UnityEnvironment
from dqn_agent import Agent
import numpy as np
from collections import deque
import matplotlib.pyplot as plt
import torch
import time
%matplotlib inline

env = UnityEnvironment(file_name="./Banana.exe")
brain_name = env.brain_names[0]
brain = env.brains[brain_name]
env_info = env.reset(train_mode=True)[brain_name]
state = env_info.vector_observations[0] # len(state) = 37
action_size = brain.vector_action_space_size # action_size = 4
agent = Agent(state_size=len(state), action_size=action_size, seed=0)
````

Then we train the agent by calling the dqn function passing in the object created before (environement, brain name and agent) plus a target of 13 as stated by the project requirements. 

Then we plot the scores obtained for each episodes to see how the agent train over time. For a less noisy result we also plot the average scores which is just the mean of the 100 last scores plotted every 100 iterations.

We can see that the agent solved the challenge of winning 13 bananas on average (over the last 100 attempts) in just over 500 iterations which is a good score.

````python
scores, avg_scores = dqn(env, brain_name, agent, 13)

# plot the scores
fig = plt.figure()
ax = fig.add_subplot(111)
plt.plot(np.arange(len(scores)), scores)
plt.ylabel('Score')
plt.xlabel('Episode #')
plt.show()

# plot the avg scores over 100 episodes
fig = plt.figure()
ax = fig.add_subplot(111)
plt.plot(np.arange(1,len(avg_scores)*100,100), avg_scores)
plt.ylabel('Avg Score (over 100 episodes)')
plt.xlabel('Episode #')
plt.show()

env.close()
````

For the sake of comparing future attempts to improve the model (for example by using the Rainbow DQN) I have also plotted the training over 2000 iterations to see how much bananas the model could get. The results is about 15 to 16 bananas maximum. Note that I used 1000 as an expected target to force the model to continue until it loses.

````python
scores, avg_scores = dqn(env, brain_name, agent, 1000)

# plot the scores
fig = plt.figure()
ax = fig.add_subplot(111)
plt.plot(np.arange(len(scores)), scores)
plt.ylabel('Score')
plt.xlabel('Episode #')
plt.show()

# plot the avg scores over 100 episodes
fig = plt.figure()
ax = fig.add_subplot(111)
plt.plot(np.arange(1,len(avg_scores)*100,100), avg_scores)
plt.ylabel('Avg Score (over 100 episodes)')
plt.xlabel('Episode #')
plt.show()

env.close()
````

The code below is used to watch the model play the game once trained. It is based on the example provided at the beginning but with a timer to slow down the replay to about 30 frame per second. A much nicer approach would be to use the elapsed time since the last frame but this approach works ok on my machine.

Note that this time the model was able to achieve a score of 17 which is very good. Below the code is a recording of that session. (Note that the video mention 600 iterations for training, but this is the model for 2000 iterations)

````python
# Load the trained agent and watch it play!
agent.qnetwork_local.load_state_dict(torch.load('checkpoint.pth'))

env_info = env.reset(train_mode=False)[brain_name] # reset the environment
state = env_info.vector_observations[0]            # get the current state
score = 0                                          # initialize the score
while True:
    action = agent.act(state)                      # select an action
    env_info = env.step(action)[brain_name]        # send the action to the environment
    next_state = env_info.vector_observations[0]   # get the next state
    reward = env_info.rewards[0]                   # get the reward
    done = env_info.local_done[0]                  # see if episode has finished
    score += reward                                # update the score
    state = next_state                             # roll over the state to next time step
    time.sleep(0.03)                               # Wait for about 30 milliseconds (30 FPS)
    if done:                                       # exit loop if episode finished
        break
    
print("Score: {}".format(score))
````
